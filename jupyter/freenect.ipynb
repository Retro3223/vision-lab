```python
%matplotlib inline
from utils import *
import freenect
```

```python
# tilt! 
ctx = freenect.init()
dev = freenect.open_device(ctx, 0)

print (ctx, dev)
state = freenect.get_tilt_state(dev)
print (freenect.get_tilt_degs(state))
```

```python
# it moves!
freenect.set_tilt_degs(dev, -10)
```

```python
# get_tilt_degs doesn't tell it to read current value
print (freenect.get_tilt_degs(state))
```

```python
# update_tilt_state does
freenect.update_tilt_state(dev)
print (freenect.get_tilt_degs(state))
```

```python
freenect.close_device(dev)
```
